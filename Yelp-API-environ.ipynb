```python
cols = list(response0['businesses'][0].keys())
cols
```

```python
data0=pd.DataFrame(response0['businesses'], columns=cols)
data1=pd.DataFrame(response1['businesses'], columns=cols)
data2=pd.DataFrame(response2['businesses'], columns=cols)
data3=pd.DataFrame(response3['businesses'], columns=cols)
data4=pd.DataFrame(response4['businesses'], columns=cols)
data5=pd.DataFrame(response5['businesses'], columns=cols)
data6=pd.DataFrame(response6['businesses'], columns=cols)
```

```python
# concat the dataframes
frames = [data0, data1, data2, data3, data4,data5, data6]
  
businesses = pd.concat(frames)
```

```python
# Collect all the reviews with respect to the businesses id.
all_reviews = []
for i in range(len(businesses)):
    res = yelp_api.reviews_query(id=businesses['id'].values[0])
    all_reviews.append(res)
```

```python
all_reviews
```

```python
cols=list(all_reviews[0]['reviews'][0].keys())
```

```python
combined_reviews=pd.DataFrame(columns=cols)

for rs in all_reviews:
    for r in rs['reviews']:
        combined_reviews = combined_reviews.append(r, ignore_index=True)
```

```python
#Save the file as csv

combined_reviews.to_csv('combined_reviews.csv', index=False)
```
